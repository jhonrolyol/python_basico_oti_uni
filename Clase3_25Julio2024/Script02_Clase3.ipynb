# Rutina de entrada
* Funcion `input`

```python
# Pidamos al usuario que ingrese su nombre
nombre = input("Ingresa tu nombre : ")
```

```python
# tipo de dato
type(nombre)
```

```python
# Pidamos al usuario que ingrese su edad en años
edad = input("Ingresa  tu edad en años")
```

```python
# tipo dato
type(edad)
```

```python
# Python es un lenguaje de programacionde tipado dinamico
edad = int(edad)
edad
```

```python
# Pidamos al usuario su estatura en metros (punto flotante)
estatura = float(input("INgresa  tu estatura (m.) : "))
estatura
```

```python
# Pidamos al usuario que ingrese su peso en kilos
peso = float(input("Ingresa tu peso en kilos (sin mentir) : "))
peso
```

```python
# Calculo del indice de masa corporal
IMC = peso/(estatura**2)
IMC
```

## Rutinas de salida
* funcion `print`

```python
# Primera forma de la funcion print
print(nombre)
print(edad)
print(IMC)
```

```python
# Segunda forma de uso de la funcion print
print("Nombre ingresado ", nombre)
print("edad : ", edad , "Estatura : ", estatura)
```

```python
# Tercera forma de uso de la funcion print :
# \n : salto de linea
# \t : tabular

print("Nombre ingresado\n", nombre)
print("edad : ", edad , "\nEstatura : ", estatura, "\nIMC ", IMC)
```

```python
# Cuarta y quinta forma de uso de la funcion print :
# docstring con % de C
print("""
Nombre ingresado : %s
Edad : %i
Estatura : %.2f
IMC : %.2f
""" %(nombre,edad, estatura, IMC))


# %s : str
# %i : int
# %f : float
```
