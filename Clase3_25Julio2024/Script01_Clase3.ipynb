## Acceso a los elementos de un dato de tipo str


```python
# Un str es un objeto con naturaleza vectorial : El acceso a sus elementos
# es por medio de indices.

# En python, los indices inician en cero
```

```python
# Primera componente (primer caracter)  de un str es el indice cero
cad1[0]
```

```python
# La segunda componente sea el indice 1
cad1[1]
```

```python
# Conclusion 1: los indices en python (leidos de izquierda a derecha) inician
# en 0 y van aumentandfo de 1 en 1
```

```python
# Notar que tambien podemos acceder a los elementos de un str de
# derecha a izquierda, en cuyo caso caso el conjunto de indices inicia
# en -1 y va disminuyendo de -1 en -1

```

```python
# Ultima componente de cad3
cad3[-1]
```

```python
# Penultima componente de cad3
cad3[-2]
```

```python
# para el acceso a porciones (slicings) de str : usamos la siguiente notacion
```

```python
# Ultimos 7 caracteres de cad2
cad2[-7:]

# Reconocer que los : al final significa que llega al final del str
```

```python
# Primeros 50 caracteres de cad3
cad3[:49+1]
```

```python
# Primeros 10 caracteres de cad2
cad2[:9+1]
```

```python
# Obtengamos los caracteres del indice 8 hasta el indice 61 de cad3
cad3[8:61+1]
```

## Operaciones con str
* + : Concatenar str
* * : repetir str

```python
# declaremos otras variables
x1 = "Clase3"
x2 = "PIT145"
```

```python
# + concatena dos o mas cadenas de str
x3 = x1 + " " +x2
x3
```

```python
# * : Repetir una str
x4 = 12 * x3
x5 = (x1 + x3)*5
```

```python
# Mostremos x4
x4
```

```python
# tipo de dato x4
type(x4)
```

```python
# numero de elementos de x4
len(x4)
```

```python
# Ultimo indice : -1
x4[-1]
```

```python
# septimo caracter de x4
x4[6]
```

```python
# + : str x str -> str

# * : int x str -> str
# * : str x int -> str
```

## Operadores logicos

```python
# Logica Proposicional

# and : y logico
# or : o logico
# not : negacion

# and y or son operaciones binarias
# and : bool x bool -> bool
# or : bool x bool -> bool

# not (operacion booleana de negacion) es una operacion unaria
# not : bool -> bool

b9 = b2 and b8
b9
```

```python
b10 = b1 or b2
b10
```

```python
# operador booleano not
not b10
```

```python
# Expresemos una operacion booleana
not ((not (b1 and b3)) or (b9 or b8))
```
