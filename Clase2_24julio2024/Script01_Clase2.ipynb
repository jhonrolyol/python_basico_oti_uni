## OPeraciones aritmeticas

```python
# + - * /
#  ** : potenciacion

# Division entera -> Algoritmo de Euclides : D = d * q +r
# q = D // d
# r = D % d

# Definamos unas 4 variables numericas
x1 = 3.14
x2 = 2.71
x3 = (x1 + x2)*(x2-0.1)
x4 = (x1+x2 +0.9)**2
```

```python
# Mostremos a x3
x3
```

```python
# Mostremos a x4
x4
```

```python
# Variables enteras
D = 2024
d = 1982

# Algoritmos de euclides : D = d *q +r
q = D //d
r = D % d
```

```python
# Mostremos el cociente
q
```

```python
# Mostremos el residuo
r
```
