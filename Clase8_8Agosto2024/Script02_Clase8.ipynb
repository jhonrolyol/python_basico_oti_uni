```python
listaDatos = [[1 , 0],[3,8]]
matNotas = np.array(listaDatos)
matNotas
```

```python
# LIsta elementos/componentes del submodulo linalg
dir(np.linalg)
```

```python
# Documentacion de la funcion inv
help(np.linalg.inv)
```

```python
# Calculemos la inversa de la matriz matNotas
matInv = np.linalg.inv(matNotas)
matInv
```

```python
# Cumple la definicion operativa de la matriz inversa
np.dot(matNotas, matInv)
```

```python
# Calculemos la inversa de mat1
Inversa_mat1 = np.linalg.inv(mat1)
Inversa_mat1
```

```python
# Verifiquemos
np.dot(mat1, Inversa_mat1)
```
