```python
def secante(f,a,b,N):
  """
  Implementacion de la estrategia explicada en clase : Metodo de la secante

  Inputs de la funcion:
  ---------------------
    f : funcion (continua en [a,b])
    a,b : floats (definen [a,b])
    N : int (Numero de elementos a construir de la sucesion)

  output:
  ------
    m_n : float (n-esimo termino de la sucesion construida)
  """
  # Verifiquemos que el producto de imagenes de a y b cumple la condicion del teorema
  if f(a)*f(b)>=0:
    print("El metodo de la biseccion falla")
    print("El intervalo no es el adecuado")
    return None

  # Consideremos que se cumple las condiciones del teorema
  a_n = a
  b_n = b

  for n in range(1,N+1):
    # Construccion de los elementos de la sucesion
    m_n = a_n -f(a_n)*(b_n - a_n)/(f(b_n) - f(a_n))
    # Calculemos la imagen mediante f de ese punto
    f_m_n = f(m_n)
    # seleccionar el subintervalo izquierdo
    if f(a_n)*f_m_n <0:
      a_n = a_n
      b_n = m_n
    # seleccionar el subintervalo derecho
    elif f(b_n)*f_m_n < 0 :
      a_n = m_n
      b_n = b_n
    # El punto sea solucion exacta
    elif f_m_n == 0:
      print("Solucion exacta")
      return m_n
    else:
      print("El metodo de la biseccion falla")
      return None
  return a_n -f(a_n)*(b_n - a_n)/(f(b_n) - f(a_n))
```

```python
# Prueba1
# Ecuacion cuadratica
def f1(x):
  return x**2 -x -1

# Raiz Aproximada
c_sombrero = secante(f1,1,2,50)

# Calculemos la aproximacion de una raiz
print("""
Secante (Aprox. a la raiz) : %.3f
f(c) : %.10f

""" %(c_sombrero, f1(c_sombrero)))
```

```python

import math as m
def f4(x):
  return m.exp(x)*(3.2*m.sin(x) - 0.5*m.cos(x))

# Aproximacion a la raiz
x1 = secante(f4, 0,2,50)
x2 = secante(f4, 2,4,50)

print("""
En [0,2] : c1 = %f
En [2,4] : c2 = %f
""" %(x1,x2))
```
