# La estructura de repeticion while

```python
contador = 0
while contador <10:
  print("python" + "\tValor actual de contador " + str(contador))
  # Actualizamos el valor de la variable contador
  # nueva valor de contador = antiguo valor de contador + 2
  contador = contador + 2
  # la anterior linea es "absolutamente necesaria", pues si no actualizamos la variable contador
  # la expresion booleana que evalua la estructura de repeticion while nunca llegaria a ser False
```

```python
# Considere el escenario donde el valor de actualizacion es un aleatorio

# Importamos el modulo random
import random as rnd

Numveces = 0
NumIter = 0
while Numveces < 20:
  NumIter = NumIter + 1
  print("Python " + "Numero de veces " + str(Numveces))
  # actualicemos Numveces con un aleatorio
  Numveces = Numveces + (rnd.randint(0,20))

print("Numero de iteraciones ", NumIter)
```

```python
# Pidamos al usuario un numero diferente de cero
# Si ingresa cero debe de pedir nuevamente el numero
#
# UPDATE al requerimiento : Solo puede equivocar 3 veces consecutivas


# Pidamos una primera vez la variable numero
numero = int(input("Ingresa un numero diferente de cero "))

# Definamos un contador de veces
NumVezError = 0

while numero == 0:
  numero = int(input("Ingresa un numero diferente de cero "))
  # Actualizar un contador de veces
  NumVezError = NumVezError +1
  if NumVezError == 2 :
    # Esta asignacion logra que el while sea False : Concluyendo asi el while
    numero = 666
    print("Se te acabaron los intentos.")

```

```python
# Ingresar un conjunto no determinado de notas y calcular el promedio
# El valor 999 indica el fin del ingreso de notas

nota = float(input("Ingrese una nota : "))

if nota != 999:
  # Variable acumuladora
  suma = nota

  # Variable contadora
  n = 1

  while nota != 999 :
    nota = float(input("Ingrese una nota : "))
    if nota != 999:
      suma =  suma + nota
      n = n + 1

  print("El promedio de notas ingresadas es : %.3f" %(suma / n))
else:
  print("No se ingresaron notas .")
```

# La estructura match

```python
# match parametro:
#   case valor1:
#     sentencia1
#     sentencia2
#   case valor2:
#     sentencia3
#     sentencia4
#   case _:
#     sentencia_por_defecto
```

```python
# pidamos un numero al usuario
num = int(input("ingresa un numero : "))

match num:
  case 1:
    print("I")
  case 2:
    print("II")
  case 3:
    print("III")
  case _:
    print("NO es 1 , 2 o 3")
```

```python
#
nombre = input("Ingresa  tu nombre : ")

match nombre:
  case "Abraham"|"abraham":
    print("Usuario existente")
  case "Elon"|"elon":
    print("Usuario Administrador")
  case "Anderson"|"anderson":
    print("Programador Senior")
  case _:
    print("Usuario Inexistente")
```

```python
#
lado_a = float(input("Ingresa el primer lado del triangulo"))

match lado_a:
  case lado_a if lado_a>0:
    print("positivo")
  case lado_a if lado_a<0:
    print("negativo")
  case _:
    print("Cero")
```
