```python
# PRimera funcion

def calculoarea(lado1, lado2, lado3):
  """
  Funcion que calcula el area de un triangulo
  """
  s = (lado1+lado2+lado3)/2
  Area = (s*(s-lado1)*(s-lado2)*(s-lado3))**(0.5)
  return Area

```

```python
# Prueba1
calculoarea(12,19,21)
```

```python
# Funcion que calcula el area y el perimetro de un rectangulo
def CalculoAreaPerimetro(base, altura):
  perimetro = 2*base + 2*altura
  area = base*altura
  return perimetro, area
```

```python
# Prueba de CalculoAreaPerimetro
b = 12.8
h = 9.81
p1, a1 = CalculoAreaPerimetro(b, h)

print("""
El perimetro es : %.3f
El area es : %.3f
""" %(p1, a1))
```

```python
# Dada una cadena de caracteres y un caracter calcular
# el numero de veces que aparece ese caracter en la
# cadena de caracteres

def CuentaCaracteres(cadena,caracter ):
  Numveces = cadena.count(caracter)
  return Numveces
```

```python
# Funcion que genera una lista de palabras aleatorias a partir de una
# cadena de caracteres

def GeneraListaPalabras(cad , num):
  """
  cad : una cadena de caracteres
  num : numero de elementos de la lista a generar
  """
  numPalabras = len(cad.split())
  listaP = []
  for c in range(num):
    indice = rnd.randint(0, numPalabras-1)
    listaP.append(cad.split()[indice])
  return listaP



```
